```python
import os
import pandas as pd
import scienceplots
import matplotlib.pyplot as plt
from model_training.measures import confusion_matrix
import numpy as np

plt.style.use(['science', 'ieee'])
CLASS_LABELS = os.listdir('../data/tsrc_cleaned/test')
```

## Test split

```python
df = pd.read_csv('../models/simple_CNN_tsrc_spectrograms_cleaned_augmented_sharpened/predictions.csv')
```

```python
y_true = df["y_true"].values
y_pred = df["y_pred"].values


cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
plt.imshow(cm, interpolation="nearest", cmap=plt.cm.Blues)
plt.title("Confusion Matrix for Test Split", pad=10, fontsize=14)
cbar = plt.colorbar()
cbar.set_label('Number of Predictions', rotation=270, labelpad=20)
tick_marks = np.arange(len(CLASS_LABELS))
plt.xticks(tick_marks, CLASS_LABELS, rotation=45)
plt.yticks(tick_marks, CLASS_LABELS)
plt.grid(True)

plt.tight_layout()
plt.ylabel("True label")
plt.xlabel("Predicted label")

plt.savefig('charts/confusion_matrix_test.png')
plt.show()
```

## Train split

```python
df = pd.read_csv('../models/simple_CNN_tsrc_spectrograms_cleaned_augmented_sharpened/predictions_train.csv')
```

```python
y_true = df["y_true"].values
y_pred = df["y_pred"].values


cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
plt.imshow(cm, interpolation="nearest", cmap=plt.cm.Blues, vmax=1000)
plt.title("Confusion Matrix for Train Split", pad=10, fontsize=14)
cbar = plt.colorbar()
cbar.set_label('Number of Predictions', rotation=270, labelpad=20)
tick_marks = np.arange(len(CLASS_LABELS))
plt.xticks(tick_marks, CLASS_LABELS, rotation=45)
plt.yticks(tick_marks, CLASS_LABELS)
plt.grid(True)

plt.tight_layout()
plt.ylabel("True label")
plt.xlabel("Predicted label")

plt.savefig('charts/confusion_matrix_train.png')
plt.show()
```
